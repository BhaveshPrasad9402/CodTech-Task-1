```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Cuisine_rating.csv")
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
print("Basic Information:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Budget'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Budget')
plt.xlabel('Budget')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Countplot for Top 10 Cuisines
plt.figure(figsize=(8, 5))
cuisine_counts = df['Cuisines'].str.split(', ').explode().value_counts().head(10)
sns.barplot(x=cuisine_counts.index, y=cuisine_counts.values, palette='muted')
plt.title('Top 10 Cuisines')
plt.xlabel('Cuisine Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='Food Rating', y='Service Rating',hue='Overall Rating',color='Blue')
plt.title("Food rating Vs Service Rating")
plt.xlabel("Food rating")
plt.ylabel("Service Rating")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df,x='Budget',color='lightblue')
plt.title("Outliers in Budget")
plt.xlabel('Budget')
plt.show()
```

```python
correlation_matrix = df[['Food Rating', 'Service Rating', 'Budget']].corr()
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
